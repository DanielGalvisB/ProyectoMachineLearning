LIMPIAR BASE DE DATOS:

```python
import pandas as pd
import numpy as np
from io import StringIO

# Leer el archivo ignorando las l√≠neas hasta @DATA
with open("/content/sample_data/Premiere-League_Stats.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

# Buscar d√≥nde empiezan los datos
data_index = next(i for i, line in enumerate(lines) if line.strip().upper() == "@DATA")
data_lines = lines[data_index + 1:]

# Crear el DataFrame
df = pd.read_csv(StringIO("".join(data_lines)), header=None)

# Asignar nombres a las columnas
df.columns = [
    "Nombre", "N√∫mero_Camiseta", "Club", "Posici√≥n", "Nacionalidad", "Edad", "Apariciones", "Victorias", "Derrotas", "Goles",
    "Goles_por_partido", "Goles_de_cabeza", "Goles_con_pie_derecho", "Goles_con_pie_izquierdo", "Penales_anotados",
    "Tiros_libres_anotados", "Disparos", "Disparos_al_arco", "Precisi√≥n_disparo_%", "Tiros_al_palo", "Ocasiones_claras_falladas",
    "Porter√≠as_a_cero", "Goles_contra", "Entradas", "√âxito_entradas_%", "Entradas_como_√∫ltimo_hombre", "Disparos_bloqueados",
    "Intercepciones", "Despejes", "Despejes_de_cabeza", "Despejes_sobre_la_l√≠nea", "Recuperaciones", "Duelos_ganados",
    "Duelos_perdidos", "Duelos_50_50_exitosos", "Duelos_a√©reos_ganados", "Duelos_a√©reos_perdidos", "Autogoles",
    "Errores_que_generan_gol", "Asistencias", "Pases", "Pases_por_partido", "Ocasiones_claras_creadas", "Centros",
    "Precisi√≥n_centros_%", "Pases_filtrados", "Pases_largos_acertados", "Atajadas", "Penales_atajados", "Despejes_con_los_pu√±os",
    "Cortes_altos", "Atajadas_con_las_manos", "Cierres_como_l√≠bero", "Saque_con_las_manos", "Saque_de_meta", "Tarjetas_amarillas",
    "Tarjetas_rojas", "Faltas", "Fueras_de_juego"
]

# Reemplazar '?' por NaN
df.replace('?', np.nan, inplace=True)

# Convertir columnas a num√©rico donde sea posible
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Eliminar columnas irrelevantes o con muchos NaN
columnas_a_eliminar = [
    "Nombre", "N√∫mero_Camiseta", "Club", "Posici√≥n", "Nacionalidad",
    "Goles_por_partido", "Penales_anotados", "Tiros_libres_anotados", "Disparos", "Disparos_al_arco",
    "Precisi√≥n_disparo_%", "Ocasiones_claras_falladas", "Porter√≠as_a_cero", "Goles_contra",
    "Entradas_como_√∫ltimo_hombre", "Despejes_sobre_la_l√≠nea", "Precisi√≥n_centros_%", "Pases_filtrados",
    "Pases_largos_acertados", "Atajadas", "Penales_atajados", "Despejes_con_los_pu√±os", "Cortes_altos",
    "Atajadas_con_las_manos", "Cierres_como_l√≠bero", "Saque_con_las_manos", "Saque_de_meta",
    "Autogoles", "Errores_que_generan_gol", "Tarjetas_rojas"
]
df.drop(columns=[col for col in columnas_a_eliminar if col in df.columns], inplace=True)

# Eliminar columnas con m√°s del 50% de NaN
umbral_nulos = len(df) * 0.5
df = df.loc[:, df.isnull().sum() < umbral_nulos]

# Eliminar filas con muchos NaN
df = df.dropna(thresh=int(df.shape[1]*0.8))

# Llenar los valores NaN restantes con la media
df = df.fillna(df.mean(numeric_only=True))

# Guardar el nuevo archivo limpio
df.to_csv("Premier_League_Stats_Limpio.txt", index=False, sep='\t')

print("‚úÖ Archivo limpio guardado como 'Premier_League_Stats_Limpio.txt'")
```

CLUSTER:

```python
# Paso 1: Importar librer√≠as necesarias
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Paso 2: Cargar el archivo limpio
df = pd.read_csv("/content/Premier_League_Stats_Limpio.txt", sep='\t')
print("‚úÖ Archivo cargado correctamente. Shape:", df.shape)

# Paso 3: Filtrar solo las columnas num√©ricas para el escalado
df_numeric = df.select_dtypes(include=[np.number])  # Selecciona solo columnas num√©ricas
print("‚úÖ Columnas num√©ricas seleccionadas:", df_numeric.columns.tolist())

# Paso 4: Escalar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_numeric)  # Escalar solo las columnas num√©ricas
print("‚úÖ Datos escalados correctamente.")

# Paso 5: Determinar n√∫mero √≥ptimo de clusters (m√©todo del codo)
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Visualizar el gr√°fico del codo
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('M√©todo del Codo')
plt.xlabel('N√∫mero de Clusters (K)')
plt.ylabel('Inercia')
plt.grid(True)
plt.show()

# Paso 6: Elegir el valor de K (seg√∫n el codo del gr√°fico)
k = 4  # Cambia este valor si ves otro mejor en el gr√°fico
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Agregar los clusters al DataFrame original
df['Cluster'] = clusters
print("‚úÖ Clusters asignados correctamente.")
print(df[['Cluster']].value_counts())

# Paso 7: Mostrar ejemplos de jugadores por cluster
for i in range(k):
    print(f"\nüîπ Cluster {i}")
    print(df[df['Cluster'] == i].head())

# Paso 8: Mostrar promedios por cluster (an√°lisis)
print("\nüìä Promedios por cluster:")
print(df.groupby('Cluster').mean(numeric_only=True))

# Paso 9: Visualizar los clusters en 2D con PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df, palette='Set2', s=100)
plt.title("Visualizaci√≥n de Clusters de Jugadores (PCA)")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.grid(True)
plt.show()

# Paso 10 (opcional): Guardar el DataFrame con clusters
# Guardar el DataFrame con los clusters
output_file = "Premier_League_Stats_Clusterizado.txt"

try:
    df.to_csv(output_file, index=False, sep='\t')
    print(f"üìÅ Archivo guardado exitosamente como '{output_file}'")

    # Verificamos que realmente existe y mostramos las primeras filas
    df_verificado = pd.read_csv(output_file, sep='\t')
    print("‚úÖ Verificaci√≥n: archivo cargado exitosamente.")
    print(df_verificado.head())
except Exception as e:
    print("‚ùå Error al guardar el archivo:", e)
```

Predicci√≥n de Goles por Regresion Lineal

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Variables predictoras (todas menos 'Goles')
X = df.drop(columns=['Goles', 'Goles_de_cabeza', 'Goles_con_pie_derecho',
                     'Goles_con_pie_izquierdo', 'PCA1', 'PCA2', 'Cluster'])  # quitamos columnas relacionadas a goles
y = df['Goles']

# Divisi√≥n entrenamiento/prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predicciones
y_pred = lr.predict(X_test)

# Evaluaci√≥n
print("Regresi√≥n Lineal:")
print("MSE:", mean_squared_error(y_test, y_pred))
print("R¬≤:", r2_score(y_test, y_pred))
```

```python
Prediccion de Goles por Arbol de Decision
```

```python
from sklearn.tree import DecisionTreeRegressor

# Modelo
tree_reg = DecisionTreeRegressor(max_depth=5, random_state=42)
tree_reg.fit(X_train, y_train)

# Predicciones
y_pred_tree = tree_reg.predict(X_test)

# Evaluaci√≥n
print("√Årbol de Decisi√≥n:")
print("MSE:", mean_squared_error(y_test, y_pred_tree))
print("R¬≤:", r2_score(y_test, y_pred_tree))
```

```python
Prediccion de Goles por Random Forest
```

```python
from sklearn.ensemble import RandomForestRegressor

# Modelo
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(X_train, y_train)

# Predicciones
y_pred_rf = rf_reg.predict(X_test)

# Evaluaci√≥n
print("Random Forest:")
print("MSE:", mean_squared_error(y_test, y_pred_rf))
print("R¬≤:", r2_score(y_test, y_pred_rf))
```

Comparacion de Algoritmos
En este experimento se utilizaron tres algoritmos de aprendizaje autom√°tico para predecir la cantidad de goles anotados por jugadores de f√∫tbol, bas√°ndose en variables estad√≠sticas de rendimiento. Los modelos utilizados fueron:

Random Forest

La siguiente tabla muestra un extracto de las predicciones generadas por cada modelo en comparaci√≥n con los goles reales:

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# 1. Preparamos los datos
X = df.drop(columns=['Goles', 'Goles_de_cabeza', 'Goles_con_pie_derecho',
                     'Goles_con_pie_izquierdo', 'PCA1', 'PCA2', 'Cluster'])
y = df['Goles']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Entrenamos los modelos
# Regresi√≥n Lineal
lr = LinearRegression()
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)

# √Årbol de Decisi√≥n
tree = DecisionTreeRegressor(max_depth=5, random_state=42)
tree.fit(X_train, y_train)
pred_tree = tree.predict(X_test)

# Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)

# 3. Creamos un DataFrame con las predicciones y los valores reales
resultados = pd.DataFrame({
    'Goles_Reales': y_test.values,
    'Predicci√≥n_LR': pred_lr,
    'Predicci√≥n_√Årbol': pred_tree,
    'Predicci√≥n_RF': pred_rf
}, index=y_test.index)

# 4. Mostramos las primeras filas para comparar
print(resultados.head(10))
```

Conclusiones

üîπ 1. Regresi√≥n Lineal

Ventajas: Es un modelo simple, r√°pido y f√°cil de interpretar.

Desventajas: Produjo valores negativos en varias predicciones (ej. -0.26, -1.02), lo cual no es realista para una variable como los goles, que no pueden ser negativos.

Precisi√≥n: En general, tiende a subestimar las cantidades de goles, especialmente en casos con goles altos (como en el √≠ndice 311 con 22 goles reales).

üîπ 2. √Årbol de Decisi√≥n
Ventajas: Captura relaciones no lineales y se adapta mejor a valores altos.

Desventajas: Muestra una tendencia a sobreestimar los goles en jugadores con valores bajos (ej. predice 6.8 goles cuando el real es 2 o 3).

Precisi√≥n: Se comporta bien en jugadores con muchos goles, pero tiende a sobredimensionar en otros casos.

üîπ 3. Random Forest
Ventajas: Es un modelo m√°s robusto y equilibrado que combina m√∫ltiples √°rboles, lo que reduce el sobreajuste.

Desempe√±o: Generalmente predice valores m√°s cercanos a los reales y con menor error en comparaci√≥n a los otros dos modelos.

Ejemplo positivo: En el caso del jugador con 22 goles reales (√≠ndice 311), Random Forest predijo 20.91, que es el valor m√°s cercano de los tres modelos.

üèÅ Conclusi√≥n Final
Entre los tres modelos evaluados:

Random Forest result√≥ ser el algoritmo m√°s efectivo, al lograr un balance entre precisi√≥n y realismo en las predicciones.

√Årbol de Decisi√≥n tuvo un desempe√±o decente, pero con tendencia a sobreestimar.

Regresi√≥n Lineal mostr√≥ limitaciones significativas, como predicciones negativas y baja precisi√≥n en valores altos.

Por tanto, para futuras predicciones de goles con este tipo de datos, se recomienda utilizar Random Forest como el modelo principal.
